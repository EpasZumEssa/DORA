PdL2 -
PBE0 def2SVP

```python
path = "/storage/mam/PdL2/PBE0/Def2SVP/"

dE = grabber(path + "dE.dat", 4)
dE_A = -813.72282739 *(2625.5)# L
dE_B = -941.62440619 *(2625.5)# PdL
dE = Ha_to_kJmol(dE)

IBSI = grabber(path + "IBSI.dat", 9)
r = grabber(path + "E_r.dat",0 ,True)
print(dE_A + dE_B)
print(dE)

# print(r)

TdS_A = 0.06042376 *(-2625.5)# L 
TdS_B = 0.06561607 *(-2625.5)# PdL
TdS_AB =  0.10318652 *(-2625.5)# PdL2

gamma_list = [0.001, 0.002, 0.005, 0.01, 0.02, 0.05, 0.1, 0.2, 0.5]
k_iter = 0.01
k_max = 10000







for i in range(len(gamma_list)):
    DORAcalc = DORA(descriptor_data=IBSI, distance_data=r, dE_data=dE, 
                    dE_A = dE_A, dE_B = dE_B, y_cleave = gamma_list[i],
                    TdS_AB = TdS_AB,TdS_A = TdS_A, TdS_B = TdS_B, D_name="IBSI",
                    infinite_distance=15,k_start=100, k_step = k_iter, max_iteration = k_max).run()
    
    
    plotter(DORAcalc, D_name = "IBSI" + str(gamma_list[i]), savepath = "/scratch/mam/DORA_Paper/figures/SI/PdL2_PBE0_SVP/")
    PrintFits(DORAcalc, savepath= "/scratch/mam/DORA_Paper/ResultsData/PdL2/water/PBE0/Def2SVP/ResultsData/IBSI_PdL2_" + str(gamma_list[i]))
    
```
